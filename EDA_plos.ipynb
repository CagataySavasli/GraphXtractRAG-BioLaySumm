```python
def merge_all_text(row):
    result = []
    result.append(row['abstract'])
    result.extend(row['sections'])
    
    text = " ".join(map(str, result))
    return text

data['all_text'] = data.apply(merge_all_text, axis=1)
```

```python
data.head()
```

```python
data.loc[1, 'all_text']
```

```python
data['word_count'] = data['all_text'].apply(lambda x: len(x.split()))
```

```python
sum(data['word_count'])/len(data)
```

```python
wc = []

for abstract in data['abstract']:
    text = " ".join(map(str, abstract))
    wc.append(len(text.split()))
```

```python
sum(wc)/len(wc)
```

```python
wc = []
for section in data['sections']:
    for section_text in section:
        text = " ".join(map(str, section_text))
        wc.append(len(text.split()))
```

```python
sum(wc)/len(wc)
```
